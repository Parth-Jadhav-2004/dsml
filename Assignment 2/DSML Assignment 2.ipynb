```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

customers = pd.read_csv("/content/Customers.csv")
customers.head()
```

```python
#Here calculated Mean of only numeric columns using df_numeric.
# Calculate the mean of all numeric columns
df_numeric = customers.select_dtypes(include=[np.number])
mean = df_numeric.mean()

# Print the mean
print(mean)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

customers = pd.read_csv("/content/Customers.csv")
```

```python
#This is to find Standard Deviation

df_numeric = customers.select_dtypes(include=[np.number])
SD = df_numeric.std()
print(SD)
```

```python
#This is to find Variance

df_numeric = customers.select_dtypes(include=[np.number])
Variance = df_numeric.var()
print(Variance)
```

```python
#Histogram of Age Column
customers = customers['Age']
plt.hist(customers, bins = 30)
plt.title('Age')
plt.show()
```
